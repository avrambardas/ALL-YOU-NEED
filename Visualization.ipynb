```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Dummy Data
x = np.arange(50)
y = x + 10 * np.random.randn(50)
print(x)
print(y)
```

```python
# plot data as connected points
plt.plot(x, y) 

# USE LINES BELOW FOR NAMING
#plt.title('A Plot') # Title of the plot
#plt.xlabel('X-axis') # X-Label
#plt.ylabel('Y-axis') # Y-Label
```

```python
# plot indevisual points
plt.scatter(x, y)
```

```python
# asign sizes and colors to each point and plot them
sizes = np.abs(np.random.randn(50)) * 100
colors = np.random.randint(0, 50, 50)
plt.scatter(x, y, s=sizes, c=colors)
```

```python
# bar plot
x = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
y = [235, 554, 582, 695, 545]
plt.bar(x, y)
```

```python
# stacked vertical bar plot
x = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
y1 = np.array([235, 554, 582, 695, 545])
y2 = np.array([100, 200, 500, 600, 800])
y3 = np.array([100, 100, 100, 100, 100])

# Making the plot for y1 list's data and plotting
p1 = plt.bar(x, y1)
# Stacking the y2 list's data at top and plotting
p2 = plt.bar(x, y2, bottom = y1)
p3 = plt.bar(x, y3, bottom = y1+y2)
# legend() is used for displaying the plot legend
plt.legend(['y1','y2','y3'])
```

```python
# horizontal barplot
x = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
y = [235, 554, 582, 695, 545]
# bar() is used for plotting a vertical bar plot
plt.barh(x, y)
```

```python
# horizontal stacked barplot
# Dummy Data
x = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
y1 = [235, 554, 582, 695, 545]
y2 = [100, 200, 500, 600, 800]
# Making the plot for y1 list's data and plotting
p1 = plt.barh(x, y1)
# Stacking the y2 list's data at top and plotting
p2 = plt.barh(x, y2, left = y1)
# legend() is used for displaying the plot legend
plt.legend(['y1','y2'])
```

```python
# pie chart
label = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
values = [235, 695, 554, 550, 545]

plt.pie(values, labels = label)
```

```python
# highlighting pie chart
label = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
values = [235, 695, 554, 550, 545]
# Defining a pie that is to explode outside of the pie chart
Explode = (0, 0.1, 0, 0,0)  # only "explode" the 2nd slice (i.e. 'Year 2')
# pie() is used for plotting a pie chart and 'explode' property is used to explode a pie
plt.pie(values, labels = label, explode = Explode)
```

```python
# histogram
x = np.random.normal(size = 1000)
plt.hist(x, density = True, bins = 30)
```

```python
# 3d line plot
x = [1, 2, 3, 4, 5] # X-coordinates
y = [1, 2, 3, 4, 5] # Y-coordinates
z = [4, 10, 20, 5, 3] # Z-ccordinates

ax = plt.axes(projection = '3d')
ax.plot3D(x, y, z)
```

```python
# 3d dots plot
x = [1, 2, 3, 4, 5] # X-coordinates
y = [1, 2, 3, 4, 5] # Y-coordinates
z = [4, 10, 20, 5, 3] # Z-ccordinates

ax = plt.axes(projection = '3d')
ax.scatter3D(x, y, z)
```

```python
# connects points with the same value
x = np.linspace(-2,2,10)
y = np.linspace(-3,3,15)

x_2d, y_2d = np.meshgrid(x,y)

z = np.exp(np.cos(4*x_2d)**2-np.sin(5*y_2d)**2)-0.8

plt.contourf(x, y, z, cmap='RdGy')
```
